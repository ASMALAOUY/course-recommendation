```python
# app.py (version "Mentor-ready")

from flask import Flask, render_template
from model import df
from recommender import recommend_by_id

app = Flask(__name__)

# Page d’accueil : liste de cours
@app.route("/")
def index():
    # On limite à 50 par exemple
    courses = df[['id', 'title', 'avg_rating', 'num_subscribers']].head(50).to_dict(orient="records")
    return render_template("index.html", courses=courses)

# Page d’un cours
@app.route("/course/<int:course_id>")
def course_page(course_id):
    course_row = df[df["id"] == course_id]
    if course_row.empty:
        return "Cours introuvable", 404

    course = course_row.iloc[0].to_dict()
    recommendations = recommend_by_id(course_id, top_n=6)

    return render_template("course.html",
                           course=course,
                           recommendations=recommendations)

if __name__ == "__main__":
    app.run(debug=True)
```
