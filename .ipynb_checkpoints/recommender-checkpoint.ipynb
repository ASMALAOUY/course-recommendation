```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
from model import df, tfidf_matrix

def recommend_by_id(course_id, top_n=5):
    idx = df.index[df['id'] == course_id]
    if len(idx) == 0:
        return []

    idx = idx[0]
    similarities = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()

    similar_idx = np.argsort(similarities)[::-1][1:top_n+1]

    results = []
    for i in similar_idx:
        results.append({
            "id": int(df.iloc[i]["id"]),
            "title": df.iloc[i]["title"],
            "rating": float(df.iloc[i].get("avg_rating", 0)),
            "subscribers": int(df.iloc[i].get("num_subscribers", 0)),
            "price": float(df.iloc[i].get("price_detail__amount", 0)),
            "url": df.iloc[i].get("url", "#"),
        })

    return results
```
